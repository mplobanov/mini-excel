```python
vws = 'aeiouy'
cns = 'bcdfghjklmnpqrstvwxz'
```

```python
from random import choice, randint
```

```python
def get_name(n: int):
    s = ''
    for i in range(n):
        yield choice(cns)
        yield choice(vws)
```

```python
"".join(list(get_name(3))).capitalize()
```

```python
def get_full_name():
    return "{} {}".format("".join(list(get_name(3))).capitalize(), "".join(list(get_name(4))).capitalize())
```

```python
get_full_name()
```

```python
jobs = [
    'Go Dev',
    'Python Dev',
    'React Dev',
    'Product Manager',
    'DevOps Engineer',
    'Team Lead',
    'HR',
    'Designer'
]
```

```python
levels = ['Junior', 'Middle', 'Senior']
```

```python
def get_job():
    job = choice(jobs)
    if job != 'Team Lead':
        job = choice(levels) + ' ' + job
    return job
```

```python
get_job()
```

```python
nums = '1234567890'
def get_tel():
    return '+79' + choice(['1', '2', '5']) + "".join([choice(nums) for i in range(8)])
```

```python
get_tel()
```
